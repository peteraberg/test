```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def readTabFile(fn):
    counts = pd.read_csv(fn, delimiter='\t', header=None, usecols=range(1,39)).values
    integrationTime = pd.read_csv(fn, delimiter='\t', header=None, usecols=[0], nrows=1).values
    data = dict(counts=counts, time=integrationTime)
    return data
```

```python
def getSampleData():
    files = ['Background.tab', 'Dark.tab', 'Drop.tab', 'SampleDark.tab']
    dataList = [readTabFile(file) for file in files]
    dataDict = dict(zip(files, dataList))
    return dataDict
```

```python
def transmittance():
    data = getSampleData()
    t = np.mean(data['Background.tab']['counts'], axis=0) - np.mean(data['Dark.tab']['counts'], axis=0)
    t = t / (data['Drop.tab']['counts'] - np.mean(data['SampleDark.tab']['counts'], axis=0))
    t = t * float(data['Drop.tab']['time']) / data['Background.tab']['time']
    t = np.log10(t)
    return t
```

```python
%%timeit
data = transmittance()
```

```python
plt.plot(data[range(0,100,10),].T)
```
